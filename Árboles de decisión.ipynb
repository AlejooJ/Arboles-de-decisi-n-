## ARBOLES DE DECISIÓN

```python
#Importamos las librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Cargamos los datos 
df = pd.read_csv(r'winequality-red.csv')
```

## Análisis exploratorio de datos (EDA)

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape) 
```

## Verificación de valores nulos y su cantidad

```python
df.isna().sum()
```

## Visualización de la distribución de la variable objetivo

```python
plt.title("Distribución de cualidad de vino")
plt.hist(df.quality)
```

```python
sns.scatterplot(x="chlorides", y="alcohol", hue="quality", data=df)
plt.title("Volumen de alcohol y sus calorías por cualidad")
plt.show()
```

```python
# Crear un boxplot para todas las características excepto 'quality'
plt.figure(figsize=(15, 8))
sns.boxplot(data=df.drop(columns=['quality']), palette="Set2")
plt.xticks(rotation=90)
plt.title('Valores atípicos en las variables')
plt.show()
```

```python
# Calcular la matriz de correlación
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables', fontsize=15)
plt.show()
```

```python
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ['quality']            
```

```python
predictors = df[predictors_col]
target = df[target_col]
```

```python
predictors
```

```python
corr = predictors.corr()
corr
```

```python
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size= 0.2, random_state=13) 
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(x_train, y_train)
```

```python
plot_tree(arbol)
```

## Predicciones 

```python
predicciones = arbol.predict(x_test)
```

```python
predicciones
```

## Examinamos matriz de confusión 

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```

```python
precision = precision_score(y_test, predicciones, average='weighted')
precision
```

```python
recall = recall_score(y_test, predicciones, average='weighted')
recall
```

```python
f1 = f1_score(y_test, predicciones, average='weighted')
f1
```

## Interpretacion 

Balance en las métricas: Las métricas de precision, recall y F1-score son muy consistentes, todas alrededor del 63%. Esto sugiere que el modelo mantiene un equilibrio saludable entre identificar correctamente los casos positivos y minimizar los falsos positivos, lo cual es muy valioso en muchos escenarios de clasificación.

Desempeño razonable en clasificación: Con una accuracy de 63.75%, el modelo muestra una buena capacidad para predecir correctamente las clases en la mayoría de los casos. En muchos contextos, este nivel de precisión puede ser suficiente, sobre todo cuando no se espera una perfección absoluta.

Modelo efectivo para patrones: El hecho de que las métricas principales sean consistentes indica que el modelo es capaz de capturar los patrones en los datos de manera efectiva, proporcionando una base sólida para análisis y decisiones posteriores.

El modelo de árbol de decisión demuestra un desempeño aceptable en la tarea de clasificación, con métricas que reflejan una identificación adecuada de las clases en el conjunto de datos. La combinación de precisión, recall y F1 Score sugiere que el modelo es efectivo para reconocer patrones, proporcionando una base sólida para el análisis del conjunto de Datos.
